```python
import numpy as np
import pandas as pd
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
import folium # map rendering library
```

Download Data

```python
csv_path='https://www.toronto.ca/ext/open_data/catalog/data_set_files/2016_neighbourhood_profiles.csv'
df = pd.read_csv(csv_path,encoding='latin1')
```

Exploring collected data

```python
df.head()
```

Colecting neighborhoods names

```python
Neighbourhoods = list(df.columns.values)
Neighbourhoods = Neighbourhoods[5:]
print(Neighbourhoods)
```

Collecting Population per neighborhood

```python
dfToronto = pd.DataFrame(index=Neighbourhoods, columns=["Population_2020","Income_2020"])
dfToronto.head()
```

```python
# Population_2016 = Population, 2020
# Income_2020 = Total income: Average amount ($)


for index, row in dfToronto.iterrows():
    dfToronto.at[index, 'Population_2020'] = df[index][2]
    dfToronto.at[index, 'Income_2020'] = df[index][2264]
    

dfToronto.sort_values('Income_2020')
```
